## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df_hist = pd.read_csv("data/hist_data.csv")

bins_q1 = 30
plt.hist(df_hist["aデータ"],bins= bins_q1,alpha = 0.5, label= 'aデータ',color=('#FF9933'))
plt.hist(df_hist["bデータ"],bins= bins_q1,alpha = 0.5, label='bデータ', color=('#33FFFF'))
plt.xlabel('温度')
plt.ylabel('度数')
plt.title('aデータとbデータの温度に関する度数分布')
plt.legend(loc = 'upper right')
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
import seaborn as sns
df_scatter = pd.read_csv("data/scatter_data.csv")

ax2 =sns.scatterplot(x='length (cm)', y='width (cm)', hue='class', data=df_scatter, palette=['red', 'blue', 'green'])
plt.grid(which='both',linestyle = '-',linewidth = 0.5)

handles2, labels2 = ax2.get_legend_handles_labels()
new_labels2 = ['class' + label if label.isdigit() else label for label in labels2]

ax2.legend(handles=handles2, labels=new_labels2, loc='upper right')

plt.title('クラス0～2のlengthとwidthに関する散布図')
plt.show()
```
